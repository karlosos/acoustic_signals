```python
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np
```

## Zadanie piewrsze

```python
n_cz = 1 # Rząd filtra czybyszew
n_b = 14 # Rząd filtra butter
fs = 44100 # Częstotliwość próbkowania
fc = 5000 # Częstotliwość graniczna
```

```python
b, a = signal.iirfilter(n_b, fc / (fs / 2), btype='lowpass', ftype='butter')
w, h = signal.freqz(b, a, 1000, fs=48000)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(w / (2*np.pi), 20 * np.log10(np.maximum(abs(h), 1e-5)))
ax.grid(which='both', axis='both')
ax.set_xlabel('Częstotliwość [Hz]')
ax.set_ylabel('Amplituda [dB]')
plt.show()
```

```python
b, a = signal.iirfilter(n_cz, fc / (fs / 2), btype='highpass', ftype='cheby2', rs=60)
w, h = signal.freqz(b, a, 1000, fs=48000)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(w / (2*np.pi), 20 * np.log10(np.maximum(abs(h), 1e-5)))
ax.grid(which='both', axis='both')
ax.set_xlabel('Częstotliwość [Hz]')
ax.set_ylabel('Amplituda [dB]')
plt.show()
```

## Zadanie drugie

```python
fo = 1000

# Oktawa (podowjenie częstotliwości)
foct = fo * 2

# Tercja wg. tabeli https://pl.wikipedia.org/wiki/Tercja_(akustyka)
fd = fo * (2 ** -1/6)
fg = fo * (2 ** 1/6)

n_cz = 3

b, a = signal.iirfilter(n_cz, fo / (fs / 2), btype='highpass', ftype='cheby2', rs=60)
wo, ho = signal.freqz(b, a, 1000, fs=48000)

b, a = signal.iirfilter(n_cz, fd / (fs / 2), btype='highpass', ftype='cheby2', rs=60)
wd, hd = signal.freqz(b, a, 1000, fs=48000)

b, a = signal.iirfilter(n_cz, fg / (fs / 2), btype='highpass', ftype='cheby2', rs=60)
wg, hg = signal.freqz(b, a, 1000, fs=48000)

b, a = signal.iirfilter(n_cz, foct / (fs / 2), btype='highpass', ftype='cheby2', rs=60)
woct, hoct = signal.freqz(b, a, 1000, fs=48000)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(wo / (2*np.pi), 20 * np.log10(np.maximum(abs(ho), 1e-5)))
ax.plot(wd / (2*np.pi), 20 * np.log10(np.maximum(abs(hd), 1e-5)))
ax.plot(wg / (2*np.pi), 20 * np.log10(np.maximum(abs(hg), 1e-5)))
ax.plot(woct / (2*np.pi), 20 * np.log10(np.maximum(abs(hoct), 1e-5)))
ax.grid(which='both', axis='both')
ax.set_xlabel('Częstotliwość [Hz]')
ax.set_ylabel('Amplituda [dB]')
plt.show()
```
